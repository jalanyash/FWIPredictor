```python
new_column_names = ['day', 'month', 'year', 'temperature', 'RH', 'Ws','Rain', 'FFMC', 'DMC','DC', 'ISI', 'BUI','FWI', 'Classes', 'Region']  
dataset.columns = new_column_names
```

```python
dataset.to_csv('Algerian_cleaned_new_dataset.csv', index=False)
```

```python
not_fire= "not fire"
new_row = {'day': 1, 'month': 6, 'year': 2012, 'temperature': 29,
           'RH': 57,'Ws': 18,'Rain': 0.0,'FFMC':65.7,'DMC': 3.4,'DC': 7.6,'ISI': 1.3,
           'BUI': 3.4,'FWI': 0.5,'Classes': not_fire,'Region': 0
}
```

```python
new_row_df = pd.DataFrame([new_row])
```

```python
dataset = pd.concat([new_row_df, dataset], ignore_index=True)
```

```python
dataset.head()
```

```python
dataset.to_csv('cleaned_algerian_final.csv', index=False)
```

```python
#Plot density for all features
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
df.hist(bins=50, figsize=(20, 15))
plt.show()
```

```python
percentage=df['Classes'].value_counts(normalize=True)*100
```

```python
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("pie Chart of classes")
plt.show()
```

```python
dataset['Classes']=np.where(dataset['Classes'].str.contains('not fire'),'not fire', 'fire')
```

```python
dftemp=dataset.loc[df['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data= dataset)
plt.ylabel('Number of fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title("Fire Analysis of sidi-Bel Regions", weight= 'bold')
```

```python
dftemp=dataset.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data= dataset)
plt.ylabel('Number of fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title("Fire Analysis of Brajaia Regions", weight= 'bold')
```

It is observed that August and July had  the most number of forest fires for both regions. And from the above plot of months, we can understand few things.
    Most of the fires happened in august and very high fires happened in  only 3 months- June, July and August.
